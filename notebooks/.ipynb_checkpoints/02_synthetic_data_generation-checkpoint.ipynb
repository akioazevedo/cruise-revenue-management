```python
import pandas as pd

df = pd.read_csv("../data/processed/cruise_pricing_synthetic_final.csv")
df.head()
```

```python
price_matrix = (
    df.groupby(["days_to_departure", "cabin_type"], as_index=False)
      .agg(price=("price", "mean"))
      .pivot(
          index="days_to_departure",
          columns="cabin_type",
          values="price"
      )
      .sort_index()
)

price_matrix.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for cabin in price_matrix.columns:
    plt.plot(
        price_matrix.index,
        price_matrix[cabin],
        label=cabin
    )

plt.xlabel("Days to Departure")
plt.ylabel("Price ($)")
plt.title("Daily Cruise Cabin Prices")
plt.legend()
plt.gca().invert_xaxis()
plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

for cabin in price_matrix.columns:
    plt.plot(
        price_matrix.index,
        price_matrix[cabin],
        label=cabin
    )

# X-axis ticks every 5 days
ticks = list(range(
    int(price_matrix.index.max()),
    int(price_matrix.index.min()) - 1,
    -5
))

plt.xticks(ticks)

plt.xlabel("Days to Departure (Countdown)")
plt.ylabel("Price ($)")
plt.title("Daily Cruise Cabin Prices")
plt.legend()
plt.gca().invert_xaxis()
plt.grid(alpha=0.3)

plt.show()
```
